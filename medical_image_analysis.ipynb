```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('xray_sample.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
plt.title('Original Medical Image')
plt.axis('off')
plt.show()
```

```python
print("Image shape (Height x Width):", image.shape)
print("Minimum pixel value:", image.min())
print("Maximum pixel value:", image.max())
print("Mean pixel value:", round(image.mean(), 2))
```

```python
plt.hist(image.ravel(), bins=256)
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Resize to 512x512
image_resized = cv2.resize(image, (512, 512))
print("Resized shape:", image_resized.shape)

```

```python
# Remove small noise / smooth image
image_clean = cv2.GaussianBlur(image_resized, (3,3), 0)
```

```python
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image_resized, cmap='gray')
plt.title('Original Resized')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image_clean, cmap='gray')
plt.title('Cleaned Image')
plt.axis('off')
plt.show()
```

```python
print("Min pixel value:", image_clean.min())
print("Max pixel value:", image_clean.max())
print("Mean pixel value:", round(image_clean.mean(), 2))

# Histogram
plt.hist(image_clean.ravel(), bins=256)
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
image_enhanced = cv2.equalizeHist(image_clean)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image_clean, cmap='gray')
plt.title('Before Enhancement')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(image_enhanced, cmap='gray')
plt.title('After Enhancement')
plt.axis('off')
plt.show()

```
